```python
from pymatgen.core import Composition, Element


def N_val(aleacio):
  formula = Composition(aleacio)
  a=[]
  b=[]
  for el, amt in formula.items():
      #print(el.full_electronic_structure)
      #print(el,amt)
      a.append(el.full_electronic_structure)
      b.append(amt)
  #print(b)
  mayor=np.zeros(len(a))
  for i in range(len(a)):
    for j in range(len(a[i])):
      if a[i][j][0]>mayor[i]:
        mayor[i]=a[i][j][0]
  #print("los mayores son:")
  #print(mayor)

  suma=np.zeros(len(a))
  for i in range(len(a)):
    for j in range(len(a[i])):
      if a[i][j][0]==mayor[i]:
        suma[i]=suma[i]+a[i][j][2]
  #print("la suma de las valencias es:")
  #print(suma)
  Val=suma
  Val_mayor=0
  Val_menor=10000
  prome=0
  for i in range(len(Val)):
    if Val[i]>Val_mayor:
      Val_mayor=Val[i]
    if Val[i]<Val_menor:
      Val_menor=Val[i]
    prome=prome+Val[i]*b[i]
  total=0
  for i in range(len(b)):
    total=total+b[i]
  prome=prome/(total)
  #print("la valencia menor es "+str(Val_menor))
  #print("la valencia mayor es "+str(Val_mayor))
  #print("el promedio es "+str(prome))
  e=np.zeros(3)
  e[0]=prome
  e[1]=Val_menor
  e[2]=Val_mayor
  return e



###CREO UNA LISTA DONDE METO LAS ALEACIONES SEPARADAS POR ELEMENTO###
elementos_sep=[]
for i in range(len(alea)):
  x=Composition(alea[i])
  elementos_sep.append(x)


###OBTENEMOS LOS VALORES QUE QUEREMOS CON LA FUNCIГ“N YA DEFINIDA###
valo_Z=np.zeros((len(alea),3))
for i in range(len(alea)):
  valo_Z[i,:]=(N_val(elementos_sep[i]))

print(valo_Z)
##OPERAMOS LOS DATOS EXTRAIDOS###
ele=len(df)
z=np.zeros((ele,3))
print(len(z[1,:]))
for j in range(len(z[:,1])):
  z[j,0]=valo_Z[j,0]
  z[j,1]=np.abs(valo_Z[j,1]-valo_Z[j,2])

###AГ‘ADIMOS AL DF LOS VALORES OPERADOS###
df ['N_val_prome']= z[:,0]
df ['N_val_dife']= z[:,1]
print(df)

```

### Machine Learning (arbol de decisiГіn & XBoost)

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
import lightgbm as lgb
from sklearn.model_selection import StratifiedKFold, cross_val_score
import numpy as np


# Cargar datos
df = pd.read_csv("mi_dataframe.csv")

X = df.iloc[:,2:25]
#print(X)
y = df["Topologico"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
#, random_state=42
# Random Forest
rf = RandomForestClassifier(
    n_estimators=500,
    random_state=42,
    class_weight="balanced"
)




rf.fit(X_train, y_train)

print("RF AUC:", roc_auc_score(y_test, rf.predict_proba(X_test)[:,1]))

# XGBoost
xgb_model = xgb.XGBClassifier(
    n_estimators=500,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric="logloss",
    random_state=42
)

xgb_model.fit(X_train, y_train)
print("XGB AUC:", roc_auc_score(y_test, xgb_model.predict_proba(X_test)[:,1]))



cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
scores = cross_val_score(xgb_model, X, y, cv=cv, scoring="roc_auc")
print("AUC promedio:", scores.mean())
print("Desviacion:", scores.std())
```

```python

texto =input("la aleaciГіn querida: ")




def propie(texto):
  #Separamos la aleciГіn
  formula=str(texto)
  print(formula)
  elementos_sep_alea=[]
  x=Composition(formula)
  elementos_sep_alea.append(x)

  #Generamos el data frame de esta aleciГіn
  df_ale = pd.DataFrame([[formula]], columns=["formula"])

  #Ordeno la formula
  comp = Composition(formula)
  formula_mp1=(comp.reduced_formula)
  print(formula_mp1)
  df_ale["formula_ordenada"] = formula_mp1


  #Numero atomico
  featurizer = ElementProperty(
    data_source=PymatgenData(),
    features=["Z"],
    stats=["mean", "minimum", "maximum"])
  valo_Z_formula=np.zeros((1,3))
  valo_Z_formula[0,:]=(featurizer.featurize(elementos_sep_alea[0]))
  Valo_Z_mean=valo_Z_formula[0,0]
  Valo_Z_range=np.abs(valo_Z_formula[0,2]-valo_Z_formula[0,1])
  Valo_Z_sum=np.abs(valo_Z_formula[0,2]+valo_Z_formula[0,1])
  df_ale ['Z_prome']= Valo_Z_mean
  df_ale ['Z_dife']= Valo_Z_range
  df_ale ['Z_sum']= Valo_Z_sum

  #Radio atomico
  featurizer = ElementProperty(
    data_source=PymatgenData(),
    features=["atomic_radius"],
    stats=["mean", "minimum", "maximum"])
  valo_Z_formula=np.zeros((1,3))
  valo_Z_formula[0,:]=(featurizer.featurize(elementos_sep_alea[0]))
  Valo_Z_mean=valo_Z_formula[0,0]
  Valo_Z_range=np.abs(valo_Z_formula[0,2]-valo_Z_formula[0,1])
  df_ale ['R_Emp_prome(Angs)']= Valo_Z_mean
  df_ale ['R_Emp_dife(Angs)']= Valo_Z_range

  #Electronegatividad
  featurizer = ElementProperty(
    data_source=PymatgenData(),
    features=["X"],
    stats=["mean", "minimum", "maximum"])
  valo_Z_formula=np.zeros((1,3))
  valo_Z_formula[0,:]=(featurizer.featurize(elementos_sep_alea[0]))
  electro_mean=valo_Z_formula[0,0]
  electro_range=np.abs(valo_Z_formula[0,1]-valo_Z_formula[0,2])
  df_ale ['Electronega_prome']= electro_mean
  df_ale ['Electronega_dife']= electro_range

  #No. Valencias
  valo_No_val=(N_val(elementos_sep_alea[0]))
  valo_No_val_prome=valo_No_val[0]
  valo_No_val_range=np.abs(valo_No_val[1]-valo_No_val[2])
  df_ale ['N_val_prome']= valo_No_val_prome
  df_ale ['N_val_dife']= valo_No_val_range

  #No. elecrones por capa
  featurizer = ElementProperty(
    data_source=MagpieData(),
    features=["NsValence", "NpValence", "NdValence", "NfValence"],
    stats=["mean"])
  valo_Z=np.zeros((1,4))
  valo_Z[0,:]=(featurizer.featurize(elementos_sep_alea[0]))
  df_ale ['No_Val_s_prome']= valo_Z[0,0]
  df_ale ['No_Val_p_prome']= valo_Z[0,1]
  df_ale ['No_Val_d_prome']= valo_Z[0,2]
  df_ale ['No_Val_f_prome']= valo_Z[0,3]

  #Tipo de red de simetria
  a=[formula_mp1]
  df_simetria = obtener_simetria_mp(a, API_KEY)
  Grup=df_simetria.iloc[0,1]
  simetria=[]
  if df_simetria.iloc[0,3]=="Cubic":
    simetria.append([1, 0, 0, 0, 0, 0, 0, 0])
  if df_simetria.iloc[0,3]=="Tetragonal":
    simetria.append([0, 1, 0, 0, 0, 0, 0, 0])
  if df_simetria.iloc[0,3]=="Orthorhombic":
    simetria.append([0, 0, 1, 0, 0, 0, 0, 0])
  if df_simetria.iloc[0,3]=="Trigonal":
    simetria.append([0, 0, 0, 1, 0, 0, 0, 0])
  if df_simetria.iloc[0,3]=="Hexagonal":
    simetria.append([0, 0, 0, 0, 1, 0, 0, 0])
  if df_simetria.iloc[0,3]=="Monoclinic":
    simetria.append([0, 0, 0, 0, 0, 1, 0, 0])
  if df_simetria.iloc[0,3]=="Triclinic":
    simetria.append([0, 0, 0, 0, 0, 0, 1, 0])
  if df_simetria.iloc[0,3]=="Desconocido":
    simetria.append([0, 0, 0, 0, 0, 0, 0, 1])
    Grup=0

  sime_array=np.zeros((1,8))
  sime_array[0,:]=simetria[0]

  #NГєmero de grupo de simetria
  df_ale ['Cubic']= sime_array[0,0]
  df_ale ['Tetragonal']= sime_array[0,1]
  df_ale ['Orthorhombic']= sime_array[0,2]
  df_ale ['Trigonal']= sime_array[0,3]
  df_ale ['Hexagonal']= sime_array[0,4]
  df_ale ['Monoclinic']= sime_array[0,5]
  df_ale ['Triclinic']= sime_array[0,6]
  df_ale ['Desconocido']= sime_array[0,7]
  df_ale ['No_Grup']= Grup

  #Es grupo de inversiГіn
  inver_grup = [
    2, 10, 11, 12, 13, 14, 15,
    47, 48, 49, 50, 51, 52,
    62, 63, 64, 65, 66,
    71, 72, 73, 74,
    83, 84, 85, 86, 87, 88,
    123, 124, 125, 126, 127,
    128, 129, 130,
    139, 140, 141, 142,
    147, 148,
    162, 163,
    166, 167,
    175, 176,
    191, 192, 193, 194,
    200, 201, 202, 203, 204, 205,
    221, 222, 223, 224, 225, 226, 227, 228, 229, 230]
  inver=0
  for m in range(len(inver_grup)):
    if Grup==inver_grup[m]:
      inver=1
  df_ale ["inver_grup"]=inver

  print(df_ale)
  return df_ale


aleacion=propie(texto)

import pandas as pd
aleacion.to_csv('aleaciГіn.csv', index=False) # index=False para no incluir el Г­ndice
print("DataFrame exportado a mi_dataframe.csv")

```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

import tensorflow as tf
from tensorflow.keras.layers import (
    Input, Dense, Embedding, Flatten, Concatenate
)
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam


df = pd.read_csv("mi_dataframe.csv")
cont_cols = [ "Z_prome", "Z_dife", "Z_sum",
    "R_Emp_prome(Angs)", "R_Emp_dife(Angs)",
    "Electronega_prome", "Electronega_dife",
    "N_val_prome", "N_val_dife",
    "No_Val_s_prome", "No_Val_p_prome",
    "No_Val_d_prome", "No_Val_f_prome",
    "inver_grup"]

X_cont = df[cont_cols].values
X_group = df["No_Grup"].values
y_2 = df["Topologico"].values


##SEPARAMOS LOS DATOS EN 80% DE ENTRENAMIENTO Y 20% DE PRUEBA
Xc_train, Xc_test, Xg_train, Xg_test, y_train, y_test = train_test_split(
    X_cont,
    X_group,
    y_2,
    test_size=0.2,
    stratify=y_2,  #teniendo igual cantidad de de topo que de no topo
    random_state=42)

#escalamos todas las categorias fisico-quimicas
scaler = StandardScaler()
Xc_train = scaler.fit_transform(Xc_train) #hacemos que cada categoria quede de la forma x=(x-media)/desvi
Xc_test  = scaler.transform(Xc_test) #los datos nuevos se rescalan con la misma media que los de entrenamiento

##### CONSTRUCCIГ“  DE LA RED ################
paciencia=100
ite=5
AUC=np.zeros((ite,2))
for i in range(ite):
  #Imput de categorias fГ­sico-quimicas
  tf.keras.backend.clear_session()
  input_cont = Input(shape=(Xc_train.shape[1],), name="cont_features")
  x_cont = Dense(20, activation="relu")(input_cont)

  #Imput del nГєmero de grupo (creamos un vector de 8 entradas que represente los 230 grupos)
  input_group = Input(shape=(1,), dtype="int32", name="space_group")
  x_group = Embedding(input_dim=231, output_dim=8, name="group_embedding")(input_group)#me genera un vector aleatorio para cada grupo y lo asigna a cada dato de entrnamiento

  x_group = Flatten()(x_group)#me quita un par de corchetes
  #x_group = Dense(8, activation="relu")(x_group)

  # unimos el resultado de la neurona de catrgorias con el vector de el grupo espacial
  x = Concatenate()([x_cont, x_group])
  x = Dense(16, activation="relu")(x)

  #neurona que asigna valor de 0 o 1 dependiendo de como aprenda
  output = Dense(1, activation="sigmoid")(x)

  #creaciГіn del modelo
  model = Model(
      inputs=[input_cont, input_group],
      outputs=output
  )
  model.compile(
      optimizer=Adam(learning_rate=1e-3),
      loss="binary_crossentropy",
      metrics=["AUC"]
  )

  # entrenamos
  history = model.fit(
      [Xc_train, Xg_train],
      y_train,
      epochs=200,
      batch_size=32,
      validation_data=([Xc_test, Xg_test], y_test),
      verbose=1,
      callbacks=[
            tf.keras.callbacks.EarlyStopping(
                monitor="val_loss",
                mode="min",
                patience=paciencia,
                restore_best_weights=True
            ) ]
    )

  loss, auc = model.evaluate([Xc_test, Xg_test], y_test)
  print("NN AUC:", auc)
  print(loss)
  AUC[i,1]=auc
  AUC[i,0]= np.argmin(history.history["val_loss"])

print(AUC)

print("AUC promedio:", np.mean(AUC[:,1]))
print("DesviaciГіn:", np.std(AUC[:,1]))
#grГЎficas
import matplotlib.pyplot as plt
loss_train = np.array(history.history['loss'])
auc_train = np.array(history.history['AUC'])
fig3, g3 = plt.subplots()
plt.plot(loss_train, label='loss_train')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
fig4, g4 = plt.subplots()
plt.plot(auc_train, label='loss_train')
plt.xlabel('Epochs')
plt.ylabel('AUC')
plt.legend()
plt.show()
```
